```python
while True:
    word = input("Enter a word (must contain at least one uppercase letter): ")
    
    if any(char.isupper() for char in word):
        print("Thank you!")
        break
    else:
        print("Error: The word must contain at least one uppercase letter.\n")
```

```python
import requests
from bs4 import BeautifulSoup
def get_wordle_solution():
    url = "https://www.tomsguide.com/news/what-is-todays-wordle-answer"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    word = soup.find("p", id="b772eb30-a3e1-4c4e-a514-78aa43892bb2")\
        .find("strong")\
        .text.strip()
    return word


wordle_solution = get_wordle_solution()
print(wordle_solution)
```
